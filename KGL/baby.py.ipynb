```python
# 데이터 이름과 값 정하기
# 년도와 총인구수 그래프 변화

y = [50428893, 50746659, 51014947, 51217803, 51361911, 51585058, 51764822, 51836239, 51769539, 51672569, 51712619] #x축을 년도로 지정
x = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

plt.title('2013~2023년의 대한민국 총 인구 변화')
plt.plot(x,y, color='skyblue', marker='o')
plt.ylim(50000000, 52000000)
plt.xlabel('년도')
plt.ylabel('인구수(만명)')

plt.show()
```

```python
# 2013, 2023 2개 년도와 구성비율 비교
labels = ['0-14세', '15-64세', '65세이상']
y13 = [14.7, 73.4, 11.9]
y23 = [11.0, 70.7, 18.2]

plt.subplot(1,2,1)
plt.title('2013년 연령 구성 비율')
plt.pie(y13, labels=labels, autopct='%.1f', shadow=True)
plt.show()

plt.subplot(1,2,2)
plt.title('2023년 연령 구성 비율')
plt.pie(y23, labels=labels, autopct='%.1f', shadow=True)
plt.show()
```
