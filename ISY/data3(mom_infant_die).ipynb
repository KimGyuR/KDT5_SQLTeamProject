```python
# data3 = '모성 및 영아 사망률 => 그 중에서 출생 전후 사망률'

import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
file = '모성_및_영아사망률_출생전후사망률_new.csv'
data = pd.read_csv(file, usecols=[1,2])
data
```

```python
data.to_csv('ISY_table3_mom_infant_die.csv')
```

```python
y_list = []
for i in range(data.shape[0]):
    y_list.append(data.iloc[i][1])
    
print(y_list)
```

```python
year_list = list(range(2013, 2022))
print(year_list)
```

```python
plt.bar(year_list, y_list, color = 'orange')
plt.title('한국의 출생전후 사망률(총출생아 천명당 사망자수)')
plt.grid(True, lw = 2, ls= '--', c= '0.5')
plt.ylim(0,4)
plt.xlabel('year')
```
