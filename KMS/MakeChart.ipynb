```python

searchlist = (
'교실수 (개)',
'교원수 (명)',
'신입원아수 (명)',
'원아수 (명)',
'유치원수 (개)',
'재취원아수 (명)',
'졸업원아수 (명)',
'직원수 (명)',
'학급수 (개)',
)
legendname = [x[0] for x in pd.read_sql("SELECT DISTINCT 행정구역0 FROM KMS_유치원", conn).values.tolist()
              # if x[0].find("전국") < 0
              ]
for wanttohighlight in ['대구', '부산' , '서울' , '강원' , '전국', '경기']:
    filename = ''
    for location in legendname:
        # 인덱스 슬라이싱을 통해서 컬럼명 조회가능
        for target in [x for x in searchlist if x.find(" ") >= 0 ][3:4]:
            result = pd.read_sql(f"SELECT * FROM KMS_유치원 where 행정구역0 = '{location}' and `설립주체별(1)` = '계' and `유치원현황별(1)` = '{target}' and  `유치원현황별(2)` = '소계' ", conn)
            rank = result['데이터'].argsort().argsort()
            pallete = sns.color_palette("flare", len(result))
            # sns.barplot(data=result, x = 'year', y = '데이터', palette=np.array(pallete[::-1])[rank])
            if location.find(wanttohighlight) >= 0:
                targetalpha = 1
                targetmarker = 'o'
                plt.title(location+ ' '+target, fontdict={'size': 16,'weight': 'bold'}, pad = 20)
                filename = location+ ' '+target + '.png'
            else:
                targetalpha = 0.1
                targetmarker = ''
            chart = sns.lineplot(data=result, x = 'year', y = '데이터',palette=np.array(pallete[::-1])[rank], alpha = targetalpha, marker=targetmarker)
            chart.set_xticks(result.year.values)
            plt.ylabel("")
            plt.xlabel("년도")
            if wanttohighlight == '전국':
                ""
            elif wanttohighlight in ('경기도','서울'):
                plt.ylim(0,100000)
                # plt.ylim(0,1000)
                pass
            elif wanttohighlight == '경기':
                plt.ylim(0,200000)
                # plt.ylim(0,3000)
            else:
                plt.ylim(0,50000)
                # plt.ylim(0,1000)
                pass
    plt.savefig(filename, dpi=250)
    plt.show()        
        # plt.show()

# plt.legend(legendname, loc='upper left', bbox_to_anchor=(1, 1))
        # plt.show()
```
